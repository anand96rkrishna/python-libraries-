**JOINT PLOT**
A join plot allows to study the relationship between 2 numeric variables. 

```
sb.jointplot(x='tip',y='total_bill',data=df)
```

```
sb.jointplot(x='tip',y='total_bill',data=df,kind ='reg')#we change the kind parameter
#curve over histogram is PDF
#Line is the best fit line
```

**PAIR PLOT**

more than two independent features

features need to be either integer or a float

```
sb.pairplot(df) # it will take only the numeric values
```

```
sb.pairplot(df,hue='day')#if we want to add a categorical value in pairplot
```

```
#value counts
df['sex'].value_counts()
```

**Categorical Plots**

Seaborn also helps us in doing the analysis on Categorical Data points. 

boxplot
violinplot
countplot
bar plot

```
#count plot
sb.countplot('sex',data=df)
#we get count of male and female
```

```

sb.countplot('day',data=df)
```

```
df['day'].value_counts()
```

```
#by default values are shown in x axis
#we can also change axis
sb.countplot(y='day',data=df)
```

```
#BAR PLOT
#same as count plot but we need to specify x and y values
## Bar plot
## Bar plot
sb.barplot(x='sex',y='total_bill',data=df)
```

```
#Violin plot 
#helps us to see both the distribution of data in terms of Kernel density estimation and the box plot
sb.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
